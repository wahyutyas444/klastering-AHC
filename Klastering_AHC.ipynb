```python
# Import library yang dibutuhkan
from matplotlib import pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
import numpy as np
```

```python
# Membuat array 2D untuk klastering
X = np.array([[1, 2], [2.5, 4.5], [2, 2], [4, 1.5], [4, 2.5]])

# Tampilkan data untuk verifikasi
print("Data untuk klastering:")
print(X)
```

```python
# Lakukan hierarchical clustering
Z = linkage(X, method='ward')  # Gunakan metode 'ward' untuk linkage
```

```python
# Plot dendrogram
plt.figure(figsize=(10, 5))  # Set ukuran figure
plt.title('Dendrogram Hierarchical Clustering')  # Judul plot
plt.xlabel('Data Points')  # Label sumbu x
plt.ylabel('Jarak')  # Label sumbu y
dendrogram(Z, leaf_rotation=90., leaf_font_size=10.)  # Buat dendrogram
plt.show()
```
